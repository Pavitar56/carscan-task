## Importing necessary libraries

```python

import pixellib       #library used for background replacement
import pandas as pd
import numpy as np
import cv2

from matplotlib import pyplot as plt


import tensorflow as tf
import keras
from pixellib.tune_bg import alter_bg
```

# Processing on image 1

```python
img1 = cv2.imread('view4.jpeg')                  #creating array of image in bgr format
```

```python
img1_rgb=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)        #converting it into rgb for plotting using plt
```

```python
plt.imshow(img1_rgb)
```

## Removing background using pixellib library and deeplabv model 

```python
change_bg = alter_bg()
change_bg.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5") #loaded the deeplabv3+ model
img1_without_background=change_bg.color_bg("view4.jpeg", colors = (255,255,255), detect = "car")

cv2.imwrite("img1_no_background.jpg", img1_without_background)      #saving no background image

img1_without_background_rgb=cv2.cvtColor(img1_without_background,cv2.COLOR_BGR2RGB)
plt.imshow(img1_without_background_rgb)
```

```python
bg1 = cv2.imread('bg1.jpg')                              #background of 1st image
bg1_rgb=cv2.cvtColor(bg1,cv2.COLOR_BGR2RGB)           
plt.imshow(bg1_rgb)
```

## Changing background of 1st image to desert background

```python
change_bg = alter_bg()
change_bg.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5") #loaded the deeplabv3+ model
img1_changed_bg=change_bg.change_bg_img(f_image_path = "img1_no_background.jpg",b_image_path = "bg1.jpg")

#f_image_path: This is the foreground image, the image which background would be changed.
#b_image_path: This is the image that will be used to change the backgroud of the foreground image.

cv2.imwrite("img1_background_change.jpg", img1_changed_bg)

img1_changed_bg_rgb=cv2.cvtColor(img1_changed_bg,cv2.COLOR_BGR2RGB)
plt.imshow(img1_changed_bg_rgb)
```
